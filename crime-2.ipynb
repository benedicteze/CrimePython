```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
 #importing the dataset 
crime= pd.read_csv('/Users/user/Downloads/git hub/Crime_Data_from_2020_to_Present.csv')
crime
```

```python
#data types and null values 
crime.info()
```

```python
# Checking the count of null values 
null=crime.isnull().sum()
null
```

```python
# Bar plot showing the distribution of null values
plt.figure(figsize=(10, 6))
null.plot.bar()
plt.title('Distribution of Null Values')
plt.xlabel('Columns')
plt.ylabel('Count')
plt.show()
```

```python
#filling all missing values with 9999
crime.fillna(9999, inplace=True)  
```

```python
# verifying that na rows and columns has been filled 
crime.isnull().sum()
```

```python
crime.head()
```

```python
#which area were the most crimes committed 
crime['AREA NAME'].unique()
```

```python
#getting the count of area names 
counts_area_name = crime['AREA NAME'].value_counts()
counts_area_name 
```

```python
#  bar chart of area with the most crimes 
plt.figure(figsize=(22, 6))
plt.bar(counts_area_name .index, counts_area_name .values)
plt.title('Area with the most crime count')
plt.xlabel('Area Names')
plt.ylabel('Count')
plt.show()
```

```python
# pie chart of counts 
colors = ['red', 'blue', 'green', 'yellow', 'orange', 'purple', 'brown', 'pink', 
          'gray', 'olive', 'teal','darkgreen', 'lightblue', 'indigo', 'salmon', 
          'lightgreen', 'lavender', 'lightpink', 'white', 'yellowgreen', 'lightblue']

plt.figure(figsize=(22, 10))
plt.pie(counts_area_name .values, labels=counts_area_name .index, autopct='%1.1f%%', colors=colors)
plt.title('Area with the most crime count as a percentage')
plt.show()
```

```python
# Of all these areas what area has the highest victim count for both genders 
plt.figure(figsize=(22, 10))
plt.title('Area names and their respective victim sex counts')
sns.countplot(x='AREA NAME', hue='Vict Sex', data=crime)
```

```python
# the area with the most crime
filtered_data_central = crime[(crime['AREA NAME'] == 'Central')]
```

```python
# Defining the age ranges for the bins
bins = [0, 20, 30, 40, 50, 60, 70, 80, 100]
```

```python
# Create a new column with the age range for each row
filtered_data_central['Vict Age'] = pd.cut(filtered_data_central['Vict Age'], bins=bins, labels=['0-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81+'])
filtered_data_central['Vict Age']
```

```python
#what is the age groups of victims in the central area 
plt.figure(figsize=(22, 10))
plt.title('Age groups of victims in the central area')
sns.countplot(x='Vict Age', hue='Vict Sex', data=filtered_data_central)
```

```python
#what is the race of victims in the central area
plt.figure(figsize=(22, 10))
plt.title('Reace of victims in the central area')
count_data = filtered_data_central.groupby(['Vict Descent', 'Vict Sex']).size().reset_index(name='count')
count_data = count_data.sort_values('count', ascending=False)
sns.barplot(x='Vict Descent', y='count', hue='Vict Sex', data=count_data)
```
